```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return (x+3)**2
def gra(x):
    return 2*(x+3)
```

```python
def gre_des(ini_x, lr, it):
    x_value = []
    y_value = []
    
    x = ini_x
    
    for i in range(it):
        x_value.append(x)
        y_value.append(func(x))
        
        x = x-lr*gra(x)
        
    return x_value, y_value

ini_x = 2
lr = 0.1
it = 100

```

```python
x_value, y_value = gre_des(ini_x, lr, it)
```

```python
local_min = x_value[-1]
print("local minima at x : ", local_min)
print("y : ", func(local_min))
```

```python
x_range = np.linspace(-10, 5)
y_range = func(x_range)
```

```python
plt.plot(x_range, y_range, label = 'Function')
plt.scatter(x_value, y_value, color = 'red', label = 'Gradient Descent')
plt.title('gradient descent')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
