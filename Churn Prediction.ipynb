```python
X = df.drop(['CustomerId', 'Exited'], axis=1)
Y = df['Exited']
```

```python
# we will convert male = 1 and female = 0
df.Gender = df.Gender.replace('Male',1)
df.Gender = df.Gender.replace('Female',0)
```

```python
# Step 3: Divide the data into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Step 4: Encode categorical variables and normalize the data
from sklearn.preprocessing import StandardScaler, OneHotEncoder
categorical_columns = ['Geography', 'Gender']
numeric_columns = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']

encoder = OneHotEncoder(sparse=False, drop='first')
X_train_encoded = encoder.fit_transform(X_train[categorical_columns])
X_test_encoded = encoder.transform(X_test[categorical_columns])

scaler = StandardScaler()
X_train_numeric = scaler.fit_transform(X_train[numeric_columns])
X_test_numeric = scaler.transform(X_test[numeric_columns])

X_train = np.hstack((X_train_encoded, X_train_numeric))
X_test = np.hstack((X_test_encoded, X_test_numeric))
```

```python
# Step 5: Initialize and build the model
# now we create our artificial neural net.
# import tensorflow
import tensorflow as tf
#import keras 
from tensorflow import keras
model = keras.Sequential([
    keras.layers.Input(shape=(X_train.shape[1],)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=20, batch_size=64, validation_split=0.1)
```

```python
# Step 6: Evaluate the model
from sklearn.metrics import accuracy_score, confusion_matrix

y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("Accuracy Score: ", accuracy)
print("Confusion Matrix:\n", conf_matrix)
```
